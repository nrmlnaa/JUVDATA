```python
# import required libraries
# for data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
```

```python
# locate the dataframe file and display the information
juv_kesalahan = pd.read_excel('Final jenis kesalahan.xlsx')
juv_kesalahan.info()
```

```python
# display all column names
print(juv_kesalahan.columns.values)
```

```python
# Rename the column name
juv_kesalahan.rename(columns = {'Jenis Kesalahan':'Type of offences',
                               'Jantina':'Gender', 'Umur':'Age',
                               'Tahun':'Year','Jumlah':'Total'}, inplace = True)
```

```python
juv_kesalahan.head()
```

```python
juv_kesalahan.tail()
```

```python
juv_kesalahan.dropna(inplace = True)
```

```python
juv_kesalahan.drop_duplicates(inplace=True)
```

```python
juv_kesalahan.isnull().values.any()
```

```python
juv_kesalahan.isnull().any()
```

```python
# display basic statistical information
juv_kesalahan.describe(include=['O'])
```

```python
# display information based of offence type (jenis kesalahan)
juv_kesalahan.groupby(by=["Type of offences"], dropna=False).count()
```

```python
juv_kesalahan['Type of offences'].unique()
```

```python
# data can be seen to have different words yet have the same meaning
# standardize the data
juv_kesalahan['Type of offences'].replace(to_replace=["Akta Kesalahan Kecil / Minor Offence Act","Akta Kesalahan Kecil"],
           value="Minor offence act" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Bersabit dengan harta benda / Property-related crimes","Harta Benda"], 
                                         value="Property-related crimes" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace="Bersabit dengan orang / People-related crimes",
                                         value="People-related crimes" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Dadah / Drug","Dadah"],value="Drug" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Judi / Gamble","Berjudi"],value="Gamble" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Lain-lain","Lain-lain ","Lain-lain / Others"],
                                         value="Others" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Lalulintas / Traffic","Lalulintas"], value="Traffic offences", inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Lari dari Sekolah Diluluskan / Escape from an Approved School","Lari dari sekolah diluluskan"]
                                         ,value="Escape from an approved school" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Melanggar Syarat Pengawasan / Infrigementof Supervision Terms","Melanggar syarat pengawasan"]
                                         ,value="Infrigementof supervision terms" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Sabit Manusia","Sabit manusia"],value="Human-related crime" ,inplace = True)

juv_kesalahan['Type of offences'].replace(to_replace=["Senjata/bahan api","Senjata/bahan api / Weapon/ Fire Arms"],
                                         value="Weapon/Fire arms" ,inplace = True)
```

```python
# display information based of offence type (jenis kesalahan)
juv_kesalahan.groupby(by=["Type of offences"], dropna=False).count()
```

```python
# display information based on jantina (gender)
juv_kesalahan.groupby(by=["Gender"], dropna=False).count()
```

```python
juv_kesalahan['Gender'].unique()
```

```python
# standardize the data
juv_kesalahan['Gender'].replace(to_replace=["L / M ","L / M","Lelaki ","Lelaki"], value="Male", inplace = True)
juv_kesalahan['Gender'].replace(to_replace=["P / F","P / F ","Perempuan"], value="Female", inplace = True)
```

```python
# display information based on jantina (gender)
juv_kesalahan.groupby(by=["Gender"], dropna=False).count()
```

```python
# display information based on age (umur)
juv_kesalahan.groupby(by=["Age"], dropna=False).count()
```

```python
juv_kesalahan['Age'].unique()
```

```python
juv_kesalahan['Age'].replace(to_replace=[" 10 - 12 tahun / age","10 - 12 tahun "], value="10 - 12 age",
                              inplace = True)
juv_kesalahan['Age'].replace(to_replace=[" 13 - 15 tahun / age","13 - 15 tahun / age","13 - 15 tahun "], 
                              value="13 - 15 age", inplace = True)
juv_kesalahan['Age'].replace(to_replace=[" 16 - 17 tahun / age","16 - 17 tahun ","16 - 17 tahun"],
                              value="16 - 17 age", inplace = True)
juv_kesalahan['Age'].replace(to_replace=[" 18 - 21 tahun / age","18 - 21 tahun "], 
                              value="18 - 21 age", inplace = True)
```

```python
# display information based on age (umur)
juv_kesalahan.groupby(by=["Age"], dropna=False).count()
```

```python
# display information based on year (tahun)
juv_kesalahan.groupby(by=["Year"], dropna=False).count()
```
