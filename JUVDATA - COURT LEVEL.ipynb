```python
# Rename the column name
juv_perintah.rename(columns = {'Peringkat Mahkamah / Court Level':'Court Level',
                               'Jantina / Gender':'Gender', 'Etnic Group / Kumpulan Etnik':'Etnic Group',
                               'Tahun / Year':'Year','Jumlah / Total':'Total'}, inplace = True)
```

```python
juv_perintah.head()
```

```python
# display information based on court level (peringkat mahkamah)
juv_perintah.groupby(by=["Court Level"], dropna=False).count()
```

```python
juv_perintah['Court Level'].unique()
```

```python
# standardize the data
juv_perintah['Court Level'].replace(to_replace=
                                           ["Mahkamah Bagi Kanak-Kanak kes tidak terkawal / Court for Children Cases beyond control",
                                            "Mahkamah Bagi Kanak-Kanak melalui pendakwaan / Court for Children Indictment",
                                            "Mahkamah Bagi Kanak-Kanak (a) melalui pendakwaan",
                                            "Mahkamah Bagi Kanak-Kanak (b) kes tidak terkawal","Mahkamah Bagi Kanak-kanak"], 
                                           value="Court for Children", inplace=True)

juv_perintah['Court Level'].replace(to_replace=["Mahkamah Majistret / Magistrate Court","Mahkamah Majistret"], 
                                           value="Magistrate Court", inplace=True)

juv_perintah['Court Level'].replace(to_replace=["Mahkamah Sesyen / Sesyen Court","Mahkamah Sesyen"], 
                                           value="Session Court", inplace=True)

juv_perintah['Court Level'].replace(to_replace=["Mahkamah Tinggi / High Court","Mahkamah Tinggi"], 
                                           value="High Court", inplace=True)

juv_perintah['Court Level'].replace(to_replace=["Lain-lain "], 
                                           value="Others", inplace=True)
```

```python
# display information based on court level (peringkat mahkamah)
juv_perintah.groupby(by=["Court Level"], dropna=False).count()
```

```python
# display information based on court level (peringkat mahkamah)
juv_perintah.groupby(by=["Gender"], dropna=False).count()
```

```python
juv_perintah['Gender'].unique()
```

```python
# standardize the data
juv_perintah['Gender'].replace(to_replace=["L / M","L","Lelaki"], value="Male", inplace=True)
juv_perintah['Gender'].replace(to_replace=["P / F","P","Perempuan"], value="Female", inplace=True)
```

```python
# display information based on gender (jantina)
juv_perintah.groupby(by=["Gender"], dropna=False).count()
```

```python
# display information based on etnicity (kumpulan etnik)
juv_perintah.groupby(by=["Etnic Group"], dropna=False).count()
```

```python
juv_perintah['Etnic Group'].unique()
```

```python
# standardize the data
juv_perintah['Etnic Group'].replace(to_replace=["Cina / Chinese","Chinese ","Cina"], value="Chinese", inplace=True)

juv_perintah['Etnic Group'].replace(to_replace=["India / Indian","Indian ","India"], value="Indian", inplace=True)

juv_perintah['Etnic Group'].replace(to_replace=["Lain-Lain / Others","Lain-lain ","Lain-lain"], value="Others",
                                       inplace=True)

juv_perintah['Etnic Group'].replace(to_replace=["Melayu / Malay","Malay ","Melayu"], value="Malay", inplace=True)

juv_perintah['Etnic Group'].replace(to_replace=["PSm ","PSm (Peninsular Natives)","Pribumi Semenanjung"], value="Peninsular Natives",
                                       inplace=True)

juv_perintah['Etnic Group'].replace(to_replace=["PSb ","PSb (Sabah Natives)","Pribumi Sabah"], value="Sabah Natives",
                                       inplace=True)

juv_perintah['Etnic Group'].replace(to_replace=["PSw (Sarawak Natives)","PSw ","Pribumi Sarawak"], value="Sarawak Natives", 
                                       inplace=True)
```

```python
# display information based on etnicity (kumpulan etnik)
juv_perintah.groupby(by=["Etnic Group"], dropna=False).count()
```

```python
# display information based on year (tahun)
juv_perintah.groupby(by=["Year"], dropna=False).count()
```
